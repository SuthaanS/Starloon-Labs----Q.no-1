```python
#Import the required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

#Dataset is loaded
data1 = pd.read_csv('/content/Data_1.txt')
data2 = pd.read_csv('/content/Data_2.txt')


signal1 = data1.iloc[:, 0].values
signal2 = data2.iloc[:, 0].values

def find_peaks_troughs(signal):
    peaks, _ = find_peaks(signal)
    troughs, _ = find_peaks(-signal)
    return peaks, troughs

peaks1, troughs1 = find_peaks_troughs(signal1)
peaks2, troughs2 = find_peaks_troughs(signal2)

plt.figure(figsize=(10, 5))
plt.plot(signal1, label='Signal 1')
plt.plot(peaks1, signal1[peaks1], 'ro', label='Peaks')
plt.plot(troughs1, signal1[troughs1], 'bo', label='Troughs')
plt.legend()
plt.title('Signal 1 with Peaks and Troughs')
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(signal2, label='Signal 2')
plt.plot(peaks2, signal2[peaks2], 'ro', label='Peaks')
plt.plot(troughs2, signal2[troughs2], 'bo', label='Troughs')
plt.legend()
plt.title('Signal 2 with Peaks and Troughs')
plt.show()

print("Signal 1 Peaks Indices:", peaks1)
print("Signal 1 Troughs Indices:", troughs1)
print("Signal 2 Peaks Indices:", peaks2)
print("Signal 2 Troughs Indices:", troughs2)
```
